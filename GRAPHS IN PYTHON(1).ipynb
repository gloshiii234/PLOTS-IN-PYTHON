```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
##Plotting first graph

x = [1,2,3,4,5]
y = [4,6,8,10,3]
plt.plot(x,y,"g^-",linewidth='5',markersize ='2',label='oranges')
plt.xlabel('No of girls',fontdict={'fontname':'Calibri','fontsize':'15'})
plt.ylabel('No of oranges taken', fontdict={'fontname':'Arial','fontsize':'10'})
plt.title("A graph showing the number of oranges taken by each boarding girl in the dormitory",
         fontdict= {'fontname':'Tahoma','fontsize':'15'})

#Line 2
x2 = np.arange(4,3,2)
plt.plot(x2, x2**2)

plt.legend()
plt.show()




```

```python
##Graph example 2

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2, 'r',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'g--')
plt.legend()
plt.show()
```

```python
##Printing the two different parts of a line separately


x = [1,2,3,4,5,6]
y = [3,9,0,3,5,6]
plt.plot(x[1:], y[1:], 'p--', label = '3x')
plt.plot(x[:5],y[:5], color = 'green',label='2x')
plt.xlabel("No of people",fontdict={'fontname': 'Calibri', 'fontsize': 10})
plt.ylabel("Books taken",fontdict={'fontname': 'Calibri', 'fontsize': 10})
plt.title("Graph showing books given out in the conference",fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xticks([0,2,4,12])
plt.yticks([1,2,3,12])
plt.legend()
plt.show()
```

```python
## Resizing graph

x = [1,2,3,4,5,6]
y = [3,9,0,3,5,6]
plt.plot(x[1:], y[1:], 'p--', label = '3x')
plt.plot(x[:5],y[:5], color = 'green',label='2x')
plt.xlabel("No of people",fontdict={'fontname': 'Calibri', 'fontsize': 10})
plt.ylabel("Books taken",fontdict={'fontname': 'Calibri', 'fontsize': 10})
plt.title("Graph showing books given out in the conference",fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

plt.xticks([0,2,4,12])
plt.yticks([1,2,3,12])
plt.legend()
plt.show()
plt.figure(figsize=(5,3),dpi=300)
plt.show()

#To save graph
plt.savefig("my graph.png")
plt.show()
```

## BARGRAPHS


```python
##Plotting a bar graph

labels = ['Age','Expenditure','Classes','Hotels']
values = [12,6,14,1]

plt.xlabel('Distribution classes for each candiddate',fontdict={'fontname':'Arial','fontsize':'13'})
plt.ylabel("Scores",fontdict={'fontname':'Comic Sans MS','fontsize':'10'})
plt.title("This is a graph showing riches of candidates")
plt.bar(labels,values,color='r')
bar = plt.bar(labels,values,color='r')
##Adding designs to bars using .set_hatch()
bar[1].set_hatch('/')
bar[0].set_hatch('|')
bar[2].set_hatch('o')
bar[3].set_hatch('*')
plt.show()
```
